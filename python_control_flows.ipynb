```python
#find the sum of even and odd numbers
lst = [1,2,3,4,5,6,7]
even_sum = 0
odd_sum = 0
for i in lst:
    if (i % 2 == 0): ##means even hai.
        even_sum = even_sum + i
    else:
        odd_sum = odd_sum + i
        
print(f"Odd sum: {odd_sum} and Even sum: {even_sum}")
```

```python
##While Condition
i = 0
even_sum = 0
odd_sum = 0
while (i <= 10):
    if (i % 2 == 0):
        even_sum = even_sum + i
    else:
        odd_sum = odd_sum + i
    i = i + 1
print(odd_sum, even_sum)
```

```python
#break: loop or function end krdeta hai
x = 1
while (x < 7):
    if (x == 4):
        break
    print(x)
    x = x + 1
```

```python
#continue: loop/function m wo step or point skip krdeta hai
x=0
while(x<7):
    x=x+1
    if x==4:
        continue
    print(x)
```
