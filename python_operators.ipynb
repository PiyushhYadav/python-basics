### Equality Operator
following operations are present in python for equality check operation-
1) is: a is b returns true if variable/identifiers a and b points to the same object.
2) is not: a  is not  b returs true if variable/identifiers a and b points to the different object.
3) == : a == b returns true if variable/identifiers a and b have the same value.
4) != : a!=b returns true if variables/identifiers a and b has different value.

```python
a = "Piyush"
b = "Sumit"
a == b
```

```python
a = "Piyush"
b = "Sumit"
print(id(a))
print(id(b))
```

```python
a = "Piyush"
b = "Piyush"
print(id(a))
print(id(b))
##this means here the variable a and b are pointing to the same memory location since their Id's are same.
```

```python
a is b
```

##### The main difference between 'is' and '==' is that 'is' checks whether the variable are pointing to the same memory or not. and '==' operator checks whether their value is equal or not.
👉 == checks value equality
👉 is checks object identity (same memory location)
Example
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True (values same)
print(a is b)  # False (different objects)

👉 Even though values are same, memory is different
 
Example where both are True
a = [1, 2, 3]
b = a

print(a == b)  # True
print(a is b)  # True

👉 Both variables point to same object
